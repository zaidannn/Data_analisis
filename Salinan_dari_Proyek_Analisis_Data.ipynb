## Visualization & Explanatory Analysis

### Pertanyaan 1:

- Kapan E-commerce-public mendapatkan review score terendah ?

```python
df_order_review = pd.read_csv("E-Commerce Public Dataset/order_reviews_dataset.csv")

df_product = df_product[['product_id']]
df_order_item = df_order_item[['order_id', 'product_id']]
df_order_review = df_order_review[['order_id', 'review_score', 'review_creation_date']]

# Melakukan join antara df_product dan df_order_item berdasarkan product_id dan order_id
df_merged = pd.merge(df_product, df_order_item, on='product_id')
df_final = pd.merge(df_merged, df_order_review, on='order_id')

selected_product_ids = df_final['product_id'].unique()[:112371]
df_final_selected = df_final[df_final['product_id'].isin(selected_product_ids)]

# Memfilter data berdasarkan tanggal review_creation_date dari bulan 08-2017 sampai 08-2018
start_date = '2017-08-01'
end_date = '2018-08-31'
df_final_selected['review_creation_date'] = pd.to_datetime(df_final_selected['review_creation_date'])
df_final_selected = df_final_selected[(df_final_selected['review_creation_date'] >= start_date) & (df_final_selected['review_creation_date'] <= end_date)]
df_final_selected['month_year'] = df_final_selected['review_creation_date'].dt.to_period('M')

# Groupby berdasarkan month_year dan menghitung jumlah review dan rata-rata review score
summary_df = df_final_selected.groupby('month_year').agg({'review_score': ['count', 'mean']}).reset_index()
summary_df.columns = ['month_year', 'review_count', 'avg_review_score']
summary_df = summary_df.sort_values('month_year')
summary_df['month_year'] = summary_df['month_year'].astype(str)

plt.figure(figsize=(16, 8))

plt.subplot(2, 1, 1)
sns.lineplot(x='month_year', y='review_count', data=summary_df, marker='o')
plt.title('Number of Reviews')
plt.xlabel('Month and Year')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)

plt.subplot(2, 1, 2)
sns.lineplot(x='month_year', y='avg_review_score', data=summary_df, marker='o', color='orange')
plt.title('Average Review Score')
plt.xlabel('Month and Year')
plt.ylabel('Average Review Score')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

- Bagaimana distribusi antara costumer dan seller disetiap state ?

```python
df_customer_unique = df_costumer.drop_duplicates(subset=['customer_state'])
customer_count_by_state = df_costumer['customer_state'].value_counts()

plt.figure(figsize=(18, 8))

plt.subplot(1, 2, 1)
customer_count_by_state.plot(kind='bar', color='red')
plt.title('Total Customer by State')
plt.xlabel('Customer State')
plt.ylabel('Total Customer')
plt.xticks(rotation=45, ha='right')

# Menghilangkan data duplikat berdasarkan kolom 'seller_state'
df_seller_unique = df_seller.drop_duplicates(subset=['seller_state'])
seller_count_by_state = df_seller['seller_state'].value_counts()

plt.subplot(1, 2, 2)
seller_count_by_state.plot(kind='bar', color='lightgreen')
plt.title('Total Sellers by State')
plt.xlabel('Seller State')
plt.ylabel('Total Sellers')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
